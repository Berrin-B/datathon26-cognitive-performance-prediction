```python
import pandas as pd
import sys
import os

# src klasörünü sisteme tanıt (Klasör yapına göre yol ayarını kontrol et)
sys.path.append(os.path.abspath("../src"))

from preprocessing import Preprocessor

# 1. Veriyi yükle
train_df = pd.read_csv('../data/raw/train.csv')
test_df = pd.read_csv('../data/raw/test_x.csv')

# 2. Preprocessor nesnesini oluştur
prepro = Preprocessor()

# 3. Sadece Train üzerinde istatistikleri öğren (Fit)
prepro.fit(train_df)

# 4. Hem Train hem Test setini dönüştür (Transform)
train_final = prepro.transform_for_eda(train_df)
test_final = prepro.transform_for_eda(test_df)

print(f"Hazır! Train boyutu: {train_final.shape}, Test boyutu: {test_final.shape}")
```
