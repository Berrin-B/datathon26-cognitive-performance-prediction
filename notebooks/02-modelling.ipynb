```python
import pandas as pd


from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import sys
import os

sys.path.append(os.path.abspath(os.path.join('..')))
from src.preprocessing import pipeline # Dosyayı aynı klasöre koyduğunu varsayıyoruz
```

```python

# 1. Veriyi Yükle
raw_data = pd.read_csv("../data/raw/train.csv")

# 2. Pipeline'ı Çalıştır (Tüm temizlik ve encoding burada biter)
df_cleaned = pipeline(raw_data)

# 3. Ayırma
X = df_cleaned.drop(['bilissel_performans_skoru'], axis=1)
y = df_cleaned['bilissel_performans_skoru']

# 4. Train-Validation Split
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.20, random_state=42)

# 5. Scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)

print("Veri hazır! Şimdi model zamanı.")
```
